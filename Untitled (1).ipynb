```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment = pd.read_csv("fear_greed_index.csv")
trades = pd.read_csv("historical_data.csv")
```

```python
print("Sentiment dataset shape:", sentiment.shape)
print("Trader dataset shape:", trades.shape)
```

```python
sentiment.head()
trades.head()
```

```python
print("Sentiment Missing Values")
print(sentiment.isnull().sum())

print("Trader Missing Values")
print(trades.isnull().sum())
```

```python
print("Sentiment duplicates:", sentiment.duplicated().sum())
print("Trader duplicates:", trades.duplicated().sum())

sentiment = sentiment.drop_duplicates()
trades = trades.drop_duplicates()
```

```python
trades['Timestamp'] = pd.to_datetime(trades['Timestamp'], unit='ms')
trades['date'] = trades['Timestamp'].dt.date
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

```python
data = pd.merge(
    trades,
    sentiment[['date','classification','value']],
    on='date',
    how='left'
)
```

```python
data.head()
```

```python
daily_pnl = data.groupby(['Account','date'])['Closed PnL'].sum().reset_index()
daily_pnl.head()
```

```python
trades_per_day = data.groupby('date').size().reset_index(name='trade_count')
trades_per_day.head()
```

```python
avg_trade_size = data.groupby('Account')['Size USD'].mean().reset_index()
avg_trade_size.head()
```

```python
long_short = data['Side'].value_counts()
print(long_short)
```

```python
pnl_sentiment = data.groupby('classification')['Closed PnL'].mean()
print(pnl_sentiment)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='classification', y='Closed PnL', data=data)
plt.title("Trader PnL During Fear vs Greed")
plt.show()
```

```python
sentiment_trades = data.groupby('classification').size()

sentiment_trades.plot(kind='bar')
plt.title("Trade Frequency vs Market Sentiment")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='classification', y='Size USD', data=data)
plt.title("Position Size During Fear vs Greed")
plt.show()
```

```python
trade_counts = data.groupby('Account').size()

threshold = trade_counts.median()

frequent_traders = trade_counts[trade_counts > threshold].index

data['trader_type'] = data['Account'].apply(
    lambda x: 'Frequent' if x in frequent_traders else 'Infrequent'
)
```

```python
data['volume_group'] = np.where(
    data['Size USD'] > data['Size USD'].median(),
    'High Volume',
    'Low Volume'
)
```

```python
pnl_by_trader = data.groupby('Account')['Closed PnL'].sum()

profitable = pnl_by_trader[pnl_by_trader > 0].index

data['profit_group'] = data['Account'].apply(
    lambda x: 'Winner' if x in profitable else 'Loser'
)
```

```python
data.groupby('classification')['Closed PnL'].describe()
```

```python
data.groupby('classification')[['Size USD','Fee']].mean()
```

During Extreme Fear markets, traders should reduce position size because volatility increases and losses are more frequent.

Strategy 2
